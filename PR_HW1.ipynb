```python
import matplotlib.pyplot as plt #importing matplotlib
import numpy as np
import cv2
from PIL import Image

img1 = plt.imread('/content/images/cat1.jpg') #reads image data
resized_image1 = cv2.resize(img1, (224, 224))
cv2.imwrite('cat1.jpg',cv2.cvtColor(resized_image1, cv2.COLOR_RGB2BGR))
img1_hsv = cv2.cvtColor(resized_image1, cv2.COLOR_BGR2HSV) 
plt.imshow(resized_image1)
plt.show()



plt.grid (axis = 'x', alpha = 0.75)
plt.grid (axis = 'y', alpha = 0.75)
plt.xlabel ('Value', fontsize = 12)
plt.ylabel ('Frequency', fontsize = 12)
plt.title ("cat1.jpg", fontsize = 12)

hist_img1 = cv2.calcHist([img1_hsv], [0,1], None, [180,256], [0,180,0,256])
cv2.normalize(hist_img1, hist_img1, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX);
_ = plt.hist(img1.ravel(), bins = 8 )
plt.show()
```
